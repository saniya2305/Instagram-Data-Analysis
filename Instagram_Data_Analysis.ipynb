```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['post_datetime'] = pd.to_datetime(df['post_datetime'])
df['hour'] = df['post_datetime'].dt.hour
df['day'] = df['post_datetime'].dt.day_name()

df.head()
```

```python
df['total_engagement'] = df['comments'] + df['shares'] + df['saves']
df[['comments','shares','saves','total_engagement']].head()
```

```python
import matplotlib.pyplot as plt

best_hour = df.groupby('hour')['total_engagement'].mean().sort_values(ascending=False)

best_hour.plot(kind='bar', figsize=(10,5))
plt.title("Best Posting Hour (Avg Engagement)")
plt.xlabel("Hour of Day")
plt.ylabel("Average Engagement")
plt.show()
```

```python
best_day = df.groupby('day')['total_engagement'].mean().sort_values(ascending=False)

best_day.plot(kind='bar', figsize=(10,5))
plt.title("Best Posting Day (Avg Engagement)")
plt.xlabel("Day")
plt.ylabel("Average Engagement")
plt.show()
```

```python
content_perf = df.groupby('content_category')['total_engagement'].mean().sort_values(ascending=False)

content_perf.plot(kind='bar', figsize=(10,5))
plt.title("Best Content Category by Engagement")
plt.xlabel("Category")
plt.ylabel("Average Engagement")
plt.show()
```

```python
df.groupby('hashtags_count')['total_engagement'].mean().head(15).plot(kind='line', figsize=(10,5))
plt.title("Hashtags Count vs Engagement")
plt.xlabel("Hashtags Count")
plt.ylabel("Average Engagement")
plt.show()
```

Insights :-

1.Audience engagement depends on content quality, timing, and relevance.

2.Posting at optimal times improves performance.

3.High-value content gains more shares and saves.

4.Strategic hashtag use increases discoverability.

5.Data-
driven content planning can significantly improve Instagram growth.
